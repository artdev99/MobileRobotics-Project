```python
def is_camera_obstructed_dark(frame, threshold=30):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    mean_brightness = np.mean(gray)
    print(f"mean {mean_brightness}")
    return mean_brightness < threshold

def is_camera_obstructed_uniform(frame, variance_threshold=100):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    variance = np.var(gray)
    print(f"var {variance}")
    return variance < variance_threshold

def is_camera_obstructed_noise(frame, noise_threshold=0.1):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
    hist = hist / hist.sum()
    max_bin = hist.max()
    print(f"max_bin {max_bin}")
    return max_bin > noise_threshold
```
